# DATA CLEANSING


#### Dropping rows of NAN value

```python
nan_df=all_data[all_data.isna().any(axis=1)]
all_data=all_data.dropna(how='all')
all_data.head()
```

#### finding 'or' and deleting 

```python
all_data=all_data[all_data['Order Date'].str[0:2]!='Or']
```

#### changing columns quantity ordered and price each to correct data type  for further analysis

```python
all_data['Quantity Ordered']=pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each']=pd.to_numeric(all_data['Price Each'])
```

### Augmenting data with additional columns for analysis

### addng month column

#### Converting type of month column from stirng to integer for furtheranalysis

```python
all_data['month']=all_data['Order Date'].str[0:2]
all_data['month']=all_data['month'].astype('int32')

```

#### ADDING SALES COLUMN


```python
all_data['sales']=all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

#### ADDNG A CITY COLUMN

```python
def city_adress(adress):
    return adress.split(',')[1]

def state(state):
    return state.split(',')[2].split(' ')[1]


all_data['city']=all_data['Purchase Address'].apply(lambda x: city_adress(x) + '('  + state(x) + ')' )
all_data
```

#### adding hour and minutes column using datetime for analysis

```python
all_data['Order Date']=pd.to_datetime(all_data['Order Date'])
all_data
```

```python
all_data['hour']=all_data['Order Date'].dt.hour
all_data['minute']=all_data['Order Date'].dt.minute
all_data
```

#### ADDING  A NEW GROUPED COLUMN FOR GROUPING THE PRODUCT SOLD TOGETHER AND TO GET INSIGHTS

```python
all_data.head()
```

```python
df=all_data[all_data['Order ID'].duplicated(keep=False)]
df['Grouped']=df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head(20)
```

```python
#dropping duplicate columns

df=df[['Order ID', 'Grouped']].drop_duplicates()
df.head()
```

### Q4:- WHICH PRODUCT ARE MOST OFTEN SOLD TOGETHER

```python

df=df[['Order ID', 'Grouped']].drop_duplicates()
df.head(20)
```

#### From the above Grouped  column we will iterate over all the item and get the final count of items Grouped mostly together

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))

for key,value in count.most_common(10):
    print(key,value)
```

#### From the above list we can conclude that iphone and lightning charging cable are mostly solved together and give reports to the company so that they can get smarter with the promotions and can increase their revenue

```python
#overlaying the two graph to give insights about the dependencies of product sold to the prices

fig ,ax1=plt.subplots()
ax2=ax1.twinx()
ax1.bar(products,quantity_ordered,color='g')
ax2.plot(products,prices,'b-')
ax1.set_xticklabels(products,rotation='vertical',size=8)
ax1.set_xlabel('product name')
ax1.set_ylabel('No of items sold')
ax2.set_ylabel('Price Each unit',color='b')
plt.show()
```
